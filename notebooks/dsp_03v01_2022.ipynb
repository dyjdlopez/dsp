In this laboratory we will be doing basic operations on signals. Specifically we will look at:
* Amplitude Scaling
* Signal Addition and Subtraction
* Signal Multiplication
* Time Shifts
* Time Reversals

```python
import numpy as np
import scipy.signal as sig
from matplotlib import pyplot as plt
```

```python
min,max = -2*np.pi, 2*np.pi
rate=50
step = np.abs(max-min)/rate
N = np.arange(min, max+1, step)

unit = lambda n : np.heaviside(n,0)
sin = lambda n,theta: np.sin(n+theta)
square = lambda n, theta: sig.square(n+theta)
exp = lambda n: np.exp(n)*unit(n)
```

```python
## Plotting out test signals
sig1 = -1*sin(N,0)
sig2 = square(N,0)

plt.figure(figsize=(8,5))
markerline, stemlines, baseline = plt.stem(
    N, sig1, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True, label='Sinusoid')
markerline.set_markerfacecolor('none')

# markerline, stemlines, baseline = plt.stem(
#     N, sig2, 
#     linefmt='grey', markerfmt='D', basefmt='black',
#     use_line_collection=True, label='Square')
# markerline.set_markerfacecolor('none')

plt.ylim(-2,2)
plt.grid()
plt.legend()
plt.show()
```

## Part 1: Amplitude Operations

### Signal Amplification
$$x_α[n]=\alpha x[n]$$

```python
### Amplitude scaling
A = 2.0

sig1 = sin(N,0)
amp_sig = A*square(N,np.pi)

plt.figure(figsize=(12,5))
markerline, stemlines, baseline = plt.stem(
    N, sig1, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True, label='Sine')
markerline.set_markerfacecolor('none')

markerline, stemlines, baseline = plt.stem(
    N, amp_sig, 
    linefmt='grey', markerfmt='D', basefmt='black',
    use_line_collection=True, label='Amplified Square')
markerline.set_markerfacecolor('none')

plt.grid()
plt.legend()
plt.show()
```

```python
### Amplitude scaling
A = -1

sig1 = sin(N,0)
iamp_sig = A*sin(N,0)

plt.figure(figsize=(12,5))
markerline, stemlines, baseline = plt.stem(
    N, sig1, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True, label='Sine')
markerline.set_markerfacecolor('none')

markerline, stemlines, baseline = plt.stem(
    N, iamp_sig, 
    linefmt='grey', markerfmt='D', basefmt='black',
    use_line_collection=True, label='Amplitude-Flipped Sine')
markerline.set_markerfacecolor('none')

plt.grid()
plt.legend()
plt.show()
```

```python
### Amplitude Attenuation
A = 0.25

sig1 = sin(N,0)
ant_sig = A*sin(N,0)

plt.figure(figsize=(12,5))
markerline, stemlines, baseline = plt.stem(
    N, sig1, 
    linefmt='orange', markerfmt='o', basefmt='black',
    use_line_collection=True, label='Sine')
markerline.set_markerfacecolor('none')

markerline, stemlines, baseline = plt.stem(
    N, ant_sig, 
    linefmt='grey', markerfmt='D', basefmt='black',
    use_line_collection=True, label='Attenuated Sine')
markerline.set_markerfacecolor('none')

plt.grid()
plt.legend()
plt.show()
```

### Signal Addition
$$z[n] = x[n]+y[n]$$

```python
sig1 = sin(N,0)
sig2 = square(N,0)
sum_sig = sig1+sig2

plt.figure(figsize=(8,5))
markerline, stemlines, baseline = plt.stem(
    N, sig2, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True, label='Square')
markerline.set_markerfacecolor('none')

# markerline, stemlines, baseline = plt.stem(
#     N, sig2, 
#     linefmt='grey', markerfmt='D', basefmt='black',
#     use_line_collection=True, label='Square')
# markerline.set_markerfacecolor('none')

# plt.plot(N,sum_sig, color='hotpink',label='sin+sqr')


plt.grid()
plt.legend()
plt.show()
```

```python
sig1 = sin(2*N,0)
sig2 = 0.01*exp(N)
sum_sig = sig1-sig2

plt.figure(figsize=(8,5))
markerline, stemlines, baseline = plt.stem(
    N, sum_sig, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True, label='sine-exp')
markerline.set_markerfacecolor('none')

# markerline, stemlines, baseline = plt.stem(
#     N, sig2, 
#     linefmt='grey', markerfmt='D', basefmt='black',
#     use_line_collection=True, label='Exponential')
# markerline.set_markerfacecolor('none')

# plt.plot(N,sum_sig, color='hotpink',label='sin+exp')


plt.grid()
plt.legend()
plt.show()
```

### Signal Subtraction
$$z[n] = x[n]-y[n]$$

```python
sig1 = sin(N,0)
sig2 = square(N,0)
sum_sig = sig1+sig2

plt.figure(figsize=(8,5))
markerline, stemlines, baseline = plt.stem(
    N, sig1-sig2, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True)
markerline.set_markerfacecolor('none')

# markerline, stemlines, baseline = plt.stem(
#     N, sig2, 
#     linefmt='grey', markerfmt='D', basefmt='black',
#     use_line_collection=True, label='Square')
# markerline.set_markerfacecolor('none')

# plt.plot(N,sum_sig, color='hotpink',label='sin-sqr')


plt.grid()
plt.legend()
plt.show()
```

```python
sig1 = sin(N,0)
sig2 = -N*unit(N)
sum_sig = sig1+sig2

plt.figure(figsize=(12,5))
markerline, stemlines, baseline = plt.stem(
    N, sig1, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True, label='Sine')
markerline.set_markerfacecolor('none')

markerline, stemlines, baseline = plt.stem(
    N, sig2, 
    linefmt='grey', markerfmt='D', basefmt='black',
    use_line_collection=True, label='Ramp')
markerline.set_markerfacecolor('none')

plt.plot(N,sum_sig, color='hotpink',label='sin-ramp')


plt.grid()
plt.legend()
plt.show()
```

### Signal Multiplication
$$z[n] = x[n]\cdot y[n]$$

```python
sig1 = sin(N,0)
sig2 = square(N,0)
prod_sig = sig1*sig2

plt.figure(figsize=(8,5))
markerline, stemlines, baseline = plt.stem(
    N, prod_sig, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True,)
markerline.set_markerfacecolor('none')

# markerline, stemlines, baseline = plt.stem(
#     N, sig2, 
#     linefmt='grey', markerfmt='D', basefmt='black',
#     use_line_collection=True, label='Square')
# markerline.set_markerfacecolor('none')

# plt.plot(N,prod_sig, color='hotpink',label='sin*sqr')

plt.ylim(-1,1.1)
plt.grid()
plt.legend()
plt.show()
```

```python
sig1 = square(3*N,0)
sig2 = 0.01*exp(N)
prod_sig = sig1*sig2

plt.figure(figsize=(12,5))
markerline, stemlines, baseline = plt.stem(
    N, sig1, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True, label='Sine')
markerline.set_markerfacecolor('none')

markerline, stemlines, baseline = plt.stem(
    N, sig2, 
    linefmt='grey', markerfmt='D', basefmt='black',
    use_line_collection=True, label='Exponential')
markerline.set_markerfacecolor('none')

plt.plot(N,prod_sig, color='hotpink',label='sin*exp')


plt.grid()
plt.legend()
plt.show()
```

# Part 2: Time Operations

### Time-Shifting (Delay)
$$x_t[n] = x[n+\Delta]$$

```python
delta = -5
sig1 = sin(N,0)
shift_sig = sin(N+delta,0)

plt.figure(figsize=(8,5))
markerline, stemlines, baseline = plt.stem(
    N, shift_sig, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True, label='Sine')
markerline.set_markerfacecolor('none')


# plt.plot(N,shift_sig, color='hotpink',label='Shifted Signal')


plt.grid()
plt.legend()
plt.show()
```

### Time-Shifting (Advance)
$$x_t[n] = x[n-\Delta]$$

```python
delta = 5
sig1 = sin(N,0)
shift_sig = sin(N-delta,0)

plt.figure(figsize=(12,5))
markerline, stemlines, baseline = plt.stem(
    N, sig1, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True, label='Sine')
markerline.set_markerfacecolor('none')


plt.plot(N,shift_sig, color='hotpink',label='Shifted Signal')


plt.grid()
plt.legend()
plt.show()
```

### Time-Scaling
$$x_s[n] = x\begin{bmatrix}\frac{n}{R}\end{bmatrix}; R>0$$

```python
r = 2
sig1 = sin(N,0)
shift_sig = sin(-N,0)

plt.figure(figsize=(8,5))
markerline, stemlines, baseline = plt.stem(
    N, sig1, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True, label='Sine')
markerline.set_markerfacecolor('none')


# plt.plot(N,shift_sig, color='hotpink',label='Scaled Signal')

plt.axvline(color='black')
plt.grid()
plt.legend()
plt.show()
```

### Time-Reversal
$$x_r[n] = x[-n]$$

```python
sig1 = N*unit(N)
flip_sig = -N*unit(-N)

plt.figure(figsize=(12,5))
markerline, stemlines, baseline = plt.stem(
    N, sig1, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True, label='Ramp')
markerline.set_markerfacecolor('none')

markerline, stemlines, baseline = plt.stem(
    N, flip_sig, 
    linefmt='grey', markerfmt='D', basefmt='black',
    use_line_collection=True, label='Inv Ramp')
markerline.set_markerfacecolor('none')


plt.grid()
plt.legend()
plt.show()
```

```python
sig1 = exp(N)
flip_sig = exp(-N)

plt.figure(figsize=(8,5))
markerline, stemlines, baseline = plt.stem(
    N, sig1, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True, label='Sine')
markerline.set_markerfacecolor('none')

# markerline, stemlines, baseline = plt.stem(
#     N, flip_sig, 
#     linefmt='grey', markerfmt='D', basefmt='black',
#     use_line_collection=True, label='Inv Ramp')
# markerline.set_markerfacecolor('none')

plt.axvline(color='black')
plt.xlim(-2*np.pi,2*np.pi)
plt.grid()
plt.legend()
plt.show()
```

```python
sig1 = sin(N,np.pi/2)
flip_sig = sin(-N,np.pi/2)

plt.figure(figsize=(12,5))
markerline, stemlines, baseline = plt.stem(
    N, sig1, 
    linefmt='blue', markerfmt='o', basefmt='black',
    use_line_collection=True, label='Cosine')
markerline.set_markerfacecolor('none')

markerline, stemlines, baseline = plt.stem(
    N, flip_sig, 
    linefmt='grey', markerfmt='D', basefmt='black',
    use_line_collection=True, label='Inv Cosine')
markerline.set_markerfacecolor('none')


plt.grid()
plt.legend()
plt.show()
```
